```python
# 3.7.7 Distribución del Assurance en los perfiles.

# Assurance da como potencialmente empleables al 83.60% de los perfiles.

# Nos referimos a potencialmente empleables cuando Assurance es >= 0.5.

freq = fichero5['Assurance'].value_counts()
values = freq.values
labels = freq.index

def gen_autopcs(values):
    def autopcs(pct):
        total = sum(values)
        val = int(round(pct*total/100.0))
        return '{p:.1f}%  ({v:d})'.format(p=pct,v=val)
    return autopcs

fig1, ax1 = plt.subplots()
ax1.pie(values, labels=labels, autopct=gen_autopcs(values))
ax1.axis('equal')

plt.title('Distribución del Scoring de empleo')
plt.show()
```
