```python
# 7.1 Creamos la columna 'Score'.

fichero3.insert(5, 'Score', 0)

fichero3
```

```python
# 7.2 Ordenamos las columnas para dejar la variable endógena.

columns = ['Gender', 'Education', 'Work Exp', 'Programming Exp', 'Score', 'Status']

fichero3 = fichero3.reindex(columns=columns)

fichero3
```

```python
# 7.3 Sustituir los NaN (Not a Number) por la opción menos favorable.

# Los busqué filtrando por columnas en Excel.

fichero3['Work Exp'] = fichero3['Work Exp'].fillna('No')

fichero3['Programming Exp'] = fichero3['Programming Exp'].fillna('No')

fichero3
```

```python
# 7.4 Esta columna se calculará en base Education, Work Exp y Prog. Exp.

# a) Los valores se han determinado en una simulación en Excel previamente.

# b) Cuando un valor es mayor o igual a 0.7 se considerará potencialmente empleable.

# c) En caso contrario, se considerará un perfil menos probable de ser contratado.

# d) El Score posee un mínimo de 0.4 y máximo de 1. Con 0.7 o más se explican el 85% 
#    en empleabilidad de la columna Status.

def suma_score(fila):

    pesos = {'Graduate': 0.5, 'Not Graduate': 0.2, 'Yes': 0.3, 'No': 0.1, 1: 0.2, 0: 0.1}
    
    resultado = fila.apply(lambda x: pesos[x]).sum()
    
    return resultado

columnas_3 = ['Education', 'Work Exp', 'Programming Exp']

fichero3['Score'] = fichero3[columnas_3].apply(suma_score, axis=1)

fichero3
```

```python
# 8.1 Creamos la columna 'Matching'.

fichero3.insert(6, 'Matching', 0)

fichero3
```

```python
# 8.2 Ordenamos las columnas para dejar de neuvo la variable endógena.

columns = ['Gender', 'Education', 'Work Exp', 'Programming Exp', 'Score', 'Matching', 'Status']

fichero3 = fichero3.reindex(columns=columns)

fichero3
```

```python
# 8.3 Ahora calculamos los valores de esta columna.

# a) Esta columna calculará si un perfil con 0.7 o más de Score es un 'Y' en Status.

# b) Un 'Y' en la columna Status significa que ha sido contratado.

# c) Con esto buscamos validar la utilidad del Score tal y como está construido.

# d) En Excel haciendo el cálculo el Score explica más del 85% de los Status.

def funcion7(fila):
    if fila['Status'] == 'Y' and fila['Score'] >= 0.7:
        return 'OK'
    else:
        return 'NOPE'

fichero3['Matching'] = fichero3.apply(funcion7, axis=1)

fichero3
```

```python
# 8.4 Normalización de la columna Score con las puntuaciones de confianza a
#     una denominada Assurance.

fichero3 = fichero3.assign(Assurance=(fichero3['Score'] - 0.4) / 0.6)

fichero3['Assurance'] = fichero3['Assurance'].round(2)

columns = ['Gender', 'Education', 'Work Exp', 'Programming Exp', 'Assurance' ,'Matching', 'Status']

fichero3 = fichero3.reindex(columns=columns)

fichero3
```
